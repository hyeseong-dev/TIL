# 1. 앞뒤가 같은 10진수 만들기 
앞뒤가 같아 거꾸로 읽어도 같은 수를 '대칭수' 또는 '회문수'라고 하죠?! 예를들면 '123454321'과 같은 수를 의미합니다. 

## 문제
10진수, 2진수, 8진수 그 어느 것으로 표현하여도 대칭수가 되는 수 중, 10진수의 10이상에서의 최솟값을 구하시오. 
> 예 9(10진수) == 1001(2진수) == 11(8진수)

- '대칭수'는 'level'과 같은 걸 말함
- 'level'과 같은 단어를 대칭이라고 함. 여기서는 수니깐 대챙수.
- 2진수와 10진수는 상대적으로 자주 들었지만 8진수는 경험이 적음
- 8진수는 사용숫자가 0~7의 8개라는 것 뿐. 2진수와 원리 자체는 동일

## 생각하는 방법

2진수에서 대칭은 가장 오른쪽 자리가 0이 되면 가장 왼쪽 자리도 0이 되어야 함. 그러나 `0으로 시작하는 수는 2진수에서 '0'외에는 없으므로 가장 오른쪽 자리가 무조건 1`이 되어야 한다는 점!
  `2진수에서 가장 오른쪽 자리가 1이라는 것은 반드시 홀수`여야 한다는 의미이기도함
즉, 홀수만 찾으면 됨. 그 다음은 10 다음잉ㄴ 11부터 순서대로 찾아 나가는 프로그램을 만들면 됨. 

```python
# 11부터 탐색 개시 
num = 11

while True:
    # 진법에 따른 변수를 선언 

    n_10 = str(num)
    n_8 = oct(num).replace("0o", "")
    n_2 = bin(num).replace("0b", "")
    
    # 앞뒤가 같은지 확인 
    if n_10 == n_10[::-1] and \
        n_8 == n_8[::-1] and \
        n_2 == n_2[::-1] :
            print(num)
            break
    num += 2
```
