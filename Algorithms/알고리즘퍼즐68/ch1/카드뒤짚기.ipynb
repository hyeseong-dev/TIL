# 카드 뒤짚기 
## 상황 
1. 1~100의 번호가 써진 100장의 카드가 순서대로 나열
2. 모든 카드는 뒷면이 위를 향한 상태로 있음
3. 한 사람(A)이 2번 카드부터 1장 간격으로 카드를 뒤짚어 나감
 - 이 경우, 2, 4, 6, ..., 100번 카드의 앞면이 나타남
4. 다른 사람(B)이 3번 카드부터 2장 간격으로 카드를 뒤짚어 나감
 - 뒷몃이 위를 향한 카드는 앞면이 보이게되고 앞면이 위를 향한 카드는 뒷면이 보이게됨
5. 이렇게 n번째의 카드부터 n-1장 간격으로 카드를 뒤집는 작업을 뒤집을 카드가 더는 없을 때까지 계속함.

## 문제 
뒤집을 카드가 더는 없을 때 뒷면이 위를 향한 카드의 번호를 모두 구해 보세요. 

> 힌트 : 왼쪽부터 순서대로 처리해 나가기만 하면 되니 간단히 구현 

## 생각하는 방법 
- 문제와 같이 카드를 차례대로 뒤집는 처리를 실행하는 것만으로 풀수 있음
- 카드 상태를 배열로 표현
- 설정 값은 앞면이 위를 보는 카드를 True, 뒤면이 위를 보는 카드를 False 


```python

N = 100
cards = [False]*N

# 2~N까지 뒤짚음
for i in range(2, N+1):
    j = i-1 
    while j < len(cards):
        cards[j] = not cards[j] # 카드가 False인 경우 not을 만나 True 바뀌어 할당.
        j += i

# 뒷면이 위를 향한 카드를 출력
for i in range(0, N):
    if not cards[i]:
        print(i + 1)
    
```

```python
for i in range(1, 100+1):
    flag = False
    for j in range(1, 100+1):
        if i % j == 0:
            flag = not flag
    if flag:
        print(i)
```
