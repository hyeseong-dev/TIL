```python

n = 13 # 입력값 

count = 1 # 동적으로 변동 되는 layer(방). 초깃값1
num = 1 # 동적으로 변동되는 벌집 번호. 초깃값1

while True:
    if num >= n: # num이 n보다 크거나 같으면 while문 종료
        break   # num은 무엇? n은 입력된 벌집 번호
    num = num + (6 * count)
    count += 1
print(count)
```
